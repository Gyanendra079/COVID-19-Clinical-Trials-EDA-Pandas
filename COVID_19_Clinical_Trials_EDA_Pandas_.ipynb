####Data Cleaning and Optimization
To clean and optimize the clinical trials dataset, dropping irrelevant or sparse columns, fixing data types, handling missing values and extracting and standardizing location information.

```python
from collections import Counter
```

```python
# Drop columns with too many missing values or irrelevant for EDA
drop_cols = ['Acronym', 'Study Documents', 'Other IDs', 'Results First Posted']
df.drop(columns=drop_cols, inplace=True)
```

```python
# Convert date columns to datetime
date_cols = ['Start Date', 'Primary Completion Date', 'Completion Date', 'First Posted', 'Last Update Posted']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# fill missing values
df['Enrollment'] = df['Enrollment'].fillna(0).astype(int)
df['Phases'] = df['Phases'].fillna('Not Reported')
df['Gender'] = df['Gender'].fillna('All')
df['Outcome Measures'] = df['Outcome Measures'].fillna('Not Reported')
```

```python
# Optimize data types
category_cols = ['Status', 'Study Results', 'Gender', 'Age', 'Phases', 'Funded Bys', 'Study Type']
for col in category_cols:
    df[col] = df[col].astype('category')

# Extract countries from Locations
def extract_countries(location):
    if pd.isnull(location):
        return []
    parts = [loc.strip() for loc in location.split('|')]
    countries = [loc.split(',')[-1].strip() for loc in parts if ',' in loc]
    return countries

df['Countries'] = df['Locations'].apply(extract_countries)

# Standardize common country name variants
country_aliases = {
    "USA": "United States",
    "U.S.A.": "United States",
    "UK": "United Kingdom",
    "Hong Kong": "China",
    "Russian Federation": "Russia",
    "Republic of Korea": "South Korea"
}
df['Countries'] = df['Countries'].apply(
    lambda countries: [country_aliases.get(c, c) for c in countries]
)

# Add total country count for each row
df['Country Count'] = df['Countries'].apply(len)

# Create a trial count per country DataFrame for EDA
all_countries = [country for sublist in df['Countries'] for country in sublist]
country_counts = pd.Series(Counter(all_countries), name='Trial Count').reset_index()
country_counts.columns = ['Country', 'Trial Count']
```

####Study Duration Analysis
To understand how long COVID-19 clinical trials typically last, we can calculate duration of each study by computing the difference between the Start Date and the Completion Date which provides insight for timeline required for executing clinical trials during a global health emergency.

```python
# Convert 'Start Date' and 'Completion Date' to datetime format
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')
df['Completion Date'] = pd.to_datetime(df['Completion Date'], errors='coerce')

# Calculate study duration in days
df['Study Duration (days)'] = (df['Completion Date'] - df['Start Date']).dt.days

# Filter out missing or negative durations
valid_durations = df['Study Duration (days)'].dropna()
valid_durations = valid_durations[valid_durations >= 0]

# Plot the distribution using seaborn
plt.figure(figsize=(10, 6))
sns.histplot(valid_durations, bins=30, kde=True, color='skyblue')     # kde option gives a smooth pdf curve
plt.title("Distribution of COVID-19 Study Durations")
plt.xlabel("Study Duration (days)")
plt.ylabel("Number of Studies")
plt.show()
```

Maximum studies show durations of few months to over a year, depending on the study phase, complexity, and recruitment size. While shorter durations are often associated with early-phase or observational studies, longer durations generally indicate late-phase trials or those involving long-term outcome tracking. The wide variability in study duration highlights both the urgency and diversity of research efforts conducted during the COVID-19 pandemic.

#### Top Countries Conducting Trials
To identify countries involved in COVID-19 clinical trials, we can visualize the geographic distribution of research efforts. This analysis helps reveal the global landscape of COVID-19 clinical research, highlighting which nations led the charge in conducting trials and we can offer insights into global healthcare responsiveness and resource allocation during the pandemic.



```python
# Checking various values for locations
df.loc[1:5,'Locations']
```

```python
import re
from collections import Counter

# Load dataset
#df = pd.read_csv("COVID clinical trials.csv")

# Drop NaNs from 'Locations' column
locations = df['Locations'].dropna()

# Extract country names as they are present in the last of locations column
country_list = []

for entry in locations:
    # Split by '|' to handle multiple locations in one entry
    parts = entry.split('|')
    for part in parts:
        # Extract country as the last item in the comma-separated string
        segments = part.strip().split(',')
        if segments:
            country = segments[-1].strip()
            country_list.append(country)

# Count occurrences of each country
country_counts = Counter(country_list)
top_countries = country_counts.most_common(10)  # Top 10 countries

# Convert to DataFrame for plotting
country_df = pd.DataFrame(top_countries, columns=['Country', 'Number of Trials'])

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=country_df, x='Number of Trials', y='Country', palette='crest')
plt.title("Top 10 Countries Conducting COVID-19 Clinical Trials")
plt.xlabel("Number of Trials")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

Countries like France, Spain, Italty are intermediate in comparison to United States which has highest recorded Clinical Trails which basically present them on higher scale as their medical research facilities have more funding which mark them as leaders of the pharmaceutical global market .

What are the most commonly studied outcome measures in COVID-19 clinical trials and identify the primary goals or endpoints researchers are measuring to evaluate the effectiveness or impact of the intervention ?

```python
# Drop null values in Outcome Measures
outcomes = df['Outcome Measures'].dropna()

# Split entries by semicolon or comma
all_outcomes = []
for entry in outcomes:
    split_items = re.split(r';|,', entry)
    cleaned = [item.strip().lower() for item in split_items if item.strip()]
    all_outcomes.extend(cleaned)                                                 # normalize and collect all outcomes

# Count frequency of each outcome measure
outcome_counter = Counter(all_outcomes)
most_common_outcomes = outcome_counter.most_common(20)

# Convert to DataFrame for better presentation
top_outcomes_df = pd.DataFrame(most_common_outcomes, columns=['Outcome Measure', 'Frequency'])

# Display result
top_outcomes_df
```

* Commonly studied outcomes include inflammatory biomarkers like IL-6 and D-dimer, clinical severity indicators such as “moderate,” “severity,” and “death,” as well as mental health aspects like anxiety and depression.      
* Hospitalization metrics and follow-up timelines (e.g., day 14, day 28) are also frequently tracked, reflecting the importance of patient recovery duration.

Clean and parse the Age column and categorize age ranges

```python
# Inspect unique values
df['Age'].unique()
```

```python
# Standardize and categorize age ranges
def clean_age(age):
    age = str(age).lower().strip()
    if 'child' in age:
        return 'Child'
    elif 'adult' in age:
        return 'Adult'
    elif 'older' in age or 'senior' in age:
        return 'Older Adult'
    elif 'all' in age:
        return 'All Ages'
    elif 'years' in age:
        # Extract numbers
        parts = age.split('to')
        try:
            low = int(parts[0].strip().split()[0])
            high = int(parts[1].strip().split()[0]) if len(parts) > 1 else low
            if high < 18:
                return 'Child'
            elif 18 <= low <= 64:
                return 'Adult'
            else:
                return 'Older Adult'
        except:
            return 'Unknown'
    else:
        return 'Unknown'

df['Age Group'] = df['Age'].apply(clean_age)
```

```python
print("\n Categorized Age Groups ")
df['Age Group'].value_counts(dropna=False)
```

```python
# Convert Enrollment to numeric
df['Enrollment'] = pd.to_numeric(df['Enrollment'], errors='coerce')

# Drop NaN enrollments
df_clean = df.dropna(subset=['Enrollment'])

# Summary statistics
print("=== Summary Statistics ===")
print(df_clean['Enrollment'].describe())
```

```python
# Categorize enrollment size into buckets
def categorize_enrollment(n):
    if n < 50:
        return 'Small (<50)'
    elif 50 <= n <= 500:
        return 'Medium (50–500)'
    else:
        return 'Large (>500)'

df_clean['Enrollment Category'] = df_clean['Enrollment'].apply(categorize_enrollment)

# Countplot for categorized enrollment
plt.figure(figsize=(8, 5))
sns.countplot(data=df_clean, x='Enrollment Category', order=['Small (<50)', 'Medium (50–500)', 'Large (>500)'], palette='Set2')
plt.title("Categorized Enrollment Sizes")
plt.xlabel("Enrollment Category")
plt.ylabel("Number of Trials")
plt.tight_layout()
plt.show()
```

 A smaller number of large-scale trials enroll more than 500 participants, contributing to a high mean and standard deviation. This indicates that while most studies are small to moderately sized, a few large-scale studies contribute disproportionately to the total enrolled population.
